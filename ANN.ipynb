## Importing library

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

## Importing DataSet

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
X = df.iloc[:, 3:-1]
```

```python
X
```

```python
y = df['Exited']
```

```python
y
```

## Encoding Catagorical Data

```python
X.dtypes
```

## Searching for the type object which has 2 categories

```python
less_than_2 = []
for col in X.columns:
    if len(X[col].unique()) == 2 and X[col].dtype=='object':
        less_than_2.append(col)
```

```python
less_than_2
```

## Applying Label Encoding

```python
X[less_than_2[0]] = X[less_than_2[0]].apply(lambda x:1 if x=='Female' else 0)
```

```python
X.dtypes
```

```python
X.head()
```

## OneHot-encoding for a data more than 2 catagories


```python
X = pd.get_dummies(X)
```

```python
X
```

```python
X.dtypes
```

## converting back to Arrays

```python
X = X.values
```

```python
X
```

```python
y
```

```python
y = y.values
```

```python
y
```

## Splitting the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_test
```

## Feuture Scaling

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
X_test
```

## Building ANN 

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=10, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=9, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=3, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

## Training ANN

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics= ['accuracy'])
```

```python
ann.fit(X_train, y_train, batch_size = 32, epochs = 500)
```

## Making Prediction

```python
ann.predict(ss.transform([[600, 0, 40, 3, 60000, 2, 1, 1, 50000, 1,0,0]])) > 0.5
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
X_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
